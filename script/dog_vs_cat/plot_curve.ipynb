```python
def read_data(data_file):
    with open(data_file, "r") as fread:
        data = [float(line.strip()) for line in fread.readlines()]
    return data

acc_v1 = read_data("acc_v1.log")
loss_v1 = read_data("loss_v1.log")
```

```python
#coding:utf-8
import codecs
import matplotlib.pyplot as plt

def get_train_points(log_file):
    with codecs.open(log_file) as fread:
        points = []
        for line in fread.readlines():
            points.append(float(line.strip()))
        return points

def visual_res(points_dict):
    colors = ['b', 'r']
    linestyles = ['-', '-.']
    labels = ['No Aug','Image Aug']
    plt.figure(figsize=(20,10))
    all_plot = len(points_dict)
    for key, val in points_dict.items():
        plt.suptitle('Image Aug VS No Aug', fontsize=16)
        if key == "loss_v1":
            plt.subplot(1, 2, 1)
            plt.plot(val, color=colors[0], linestyle=linestyles[0], label=labels[0])
        if key == "loss_v2":
            plt.subplot(1, 2, 1)
            plt.plot(val, color=colors[1], linestyle=linestyles[1], label=labels[1])
            plt.title("Val loss")
            plt.ylabel("Val loss")
        if key == "acc_v1":
            plt.subplot(1, 2, 2)
            plt.plot(val, color=colors[0], linestyle=linestyles[0], label=labels[0])
        if key == "acc_v2":
            plt.subplot(1, 2, 2)
            plt.plot(val, color=colors[1], linestyle=linestyles[1], label=labels[1])
            plt.title("Val acc")
            plt.ylabel("Val acc")
        plt.legend()
    plt.show()


if __name__ == "__main__":
    loss_v1 = get_train_points("./loss_v1.log")
    loss_v2 = get_train_points("./loss_v2.log")
    acc_v1 = get_train_points("./acc_v1.log")
    acc_v2 = get_train_points("./acc_v2.log")

    data = {
        "loss_v1": loss_v1,
        "loss_v2": loss_v2,
        "acc_v1": acc_v1,
        "acc_v2": acc_v2
    }
    visual_res(data)



```
